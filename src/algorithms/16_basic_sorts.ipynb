## Bubble sort

Bubble sort is a sort algorithm which start taking the first value in a list an then compares with respect the nex following item.
1. If the new item value is less than my evaluated item, then I switch the position (move my greater value to the right).
2. I do the same question all over through the entire list until I reach certain value which is greater than my item and, when that happens: 
3. I took that new value (which is greater) and start to ask the next following item is less or greater, and repat than example continously.

For example:

1. 

2. 

3. 

4. 

5. 

6. 

7. 

and finally when there's no more items to compare with, my largest item is sorted out.

8. 


Then I just need to repeat the procedure all over again (for current first item and so on) and eventually I will ended with a sorted list.



 The time complexity of a bubble sort implementation is O(n^2). This means that the time it takes to sort the list grows quadratically with the number of elements in the list.

 The space complexity of a bubble sort algorithm is generally considered to be O(1), or constant space. This is because the algorithm sorts the elements in place, without requiring additional memory proportional to the input size. Bubble sort compares and swaps elements within the existing array, but it doesn't use extra data structures or allocate additional memory based on the size of the input.

```python
# Bubble sort code

my_list: list = [4, 2, 6, 5, 1, 3]


def bubble_sort(list_:list) -> list:
    for i in range(len(list_) - 1, 0, -1):
        for j in range(i):
            if list_[j] > list_[j+1]:
                temp = list_[j]
                list_[j] = list_[j + 1]
                list_[j+1] = temp

    return list_



print(f'Pre bubble sort: {my_list}')
print(f'After bubble sort: {bubble_sort(my_list)}')

```

# Selection sort

Selection sort is an algorithm that has an inverse philosphy with respect the bubble sort.

In this method, I will be looking for the minimun value (in fact, I will use a temporary variable to store the minimum value index).

I will start with the first position and compare it with the following item, if the following item value if less than my current index, then I will store the new minimum value index in the temporary variable and take this position value as the new minimun value to compare with the rest of the list. I will use this methodology over the entire list.

---


**The selection sort algorithm has a time complexity of O(n^2) and a space complexity of O(1).**

Time Complexity

The time complexity of an algorithm measures how long it takes the algorithm to run as the input size increases. In the case of selection sort, the algorithm performs n^2 comparisons in the worst-case scenario. This is because, in each iteration of the outer loop, the algorithm searches the entire unsorted portion of the array for the minimum element. This search requires n comparisons, and there are a total of n iterations of the outer loop. Therefore, the worst-case time complexity of selection sort is O(n^2).

The average-case and best-case time complexities of selection sort are also O(n^2). This is because the algorithm always performs n^2 comparisons, regardless of the order of the input array.

Space Complexity

The space complexity of an algorithm measures how much memory the algorithm uses as the input size increases. In the case of selection sort, the algorithm only uses a constant amount of memory, regardless of the input size. This is because the algorithm only needs to store a few temporary variables, such as the minimum element and its index. Therefore, the space complexity of selection sort is O(1).

```python
# Selection sort - Code

my_list: list = [4, 2, 6, 5, 1, 3]


def selection_sort(list_:list) -> list:
    for i in range(len(my_list)-1):

        min_index = i
        for j in range(i+1, len(my_list)):
            if my_list[j] < my_list[min_index]:
                min_index = j

        if i != min_index:
            temp = my_list[i]
            my_list[i] = my_list[min_index]
            my_list[min_index] = temp

    return my_list


print(f'Pre selection sort: {my_list}')
print(f'After selection sort: {bubble_sort(my_list)}')
```

# Insertion sort

Insertion sort is a simple sorting algorithm that works by iterating over the input list and inserting each element into its correct position in the sorted portion of the list. It is similar to how you might sort a hand of cards.

- Time Complexity (Worst-case, Average-case)	O(n^2)
- Space Complexity	O(1)


```python
def insertion_sort(my_list):
    for i in range(1, len(my_list)):
        temp = my_list[i]
        j = i-1
        while temp < my_list[j] and j > -1:
            my_list[j+1] = my_list[j] 
            my_list[j] = temp
            j -= 1
    return my_list





print(insertion_sort([4,2,6,5,1,3]))

```
