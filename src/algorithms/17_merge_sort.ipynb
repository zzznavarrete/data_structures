# Merge sort

The way how merge sort works is as follows:
1. Break the list in such a way that from the original list, it converts each element into an unique list. 




2. And a list of one item only, is by definition, sorted. Then, it starts to merge by two elements (two lists of 1 item each) and sorted. Then the number of list will be reduced by the half and each new list of two items will be sorted.



3. Then it repeats the step 2 but now with the lists of two items each, to converges into sorted sets of lists of 4 items each. Eventually the algorithm does this until it reach only one list sorted.










## Merge function

The helper function that this method relay on is 'merge function'.

The mission of this function is two take 2 sorted list and takes the values of them an put it into a new merged sorted list. 

The way of how this method works is essentially comparing the first element of the list A with the first element of the list B and ask which one is smaller. The one that is smaller gets into positioned into the new merged list. Then it does this repeatedly until there is only one merged list built by this method.








### ...





```python
# Merge sort -- CODE

def merge(list1:list, list2:list): 
    """
    """
    combined:list = []
    i:int = 0
    j:int = 0

    # while loop to iterate only if there are items in both lists
    while i < len(list1) and j < len(list2): 
        if list1[i] < list2[j]:
            combined.append(list1[i])
            i += 1
        else:
            combined.append(list2[j])
            j += 1
    while i < len(list1):
        combined.append(list1[i])
        i += 1
    while j < len(list2):
        combined.append(list2[j])
        j += 1
    return combined 


list1_, list2_ = [1,2,7,8], [3,4,5,6]


print(merge(list1_, list2_))
```

Now that we have the code that create a sorted list from another sorted lists, we're going to use recursion to approach to iteratively breaks the original unsorted lists into smaller lists, which is the first step of this algorithm.

For use recursion we need to ensure the following two principles:
1. Do the same thing over and over,
2. Each iteration must make the problem smaller.

Given that our use case accomplish those requirements, we're able to use recursion.


And we're going to use recursion until we reach the _base case_. And the _base case_ would be once we got lists of length == 1.

When we reach the _base case_ is when we're going to use the helper method _merge_ to sort our lists.

```python
def merge_sort(my_list:list) -> list:
    if len(my_list) == 1:
        return my_list
    mid_index:int = int(len(my_list) / 2)
    left = merge_sort(my_list[:mid_index])
    right = merge_sort(my_list[mid_index:])

    return merge(left, right)


original_list = [3,1,4,2]

sorted_list = merge_sort(original_list)

print(f'Original list = {original_list}')
print(f'Sorted list = {sorted_list}')
```

## Big O

### Space complexity
Given that the methods stores a number of elements equal of the number of elements in the list, then the space complexity of merge sort is $O(n)$.

### Time complexity
The time complexity of this algorithm is composed by the combination of breaking the lists and sorted them.

The first routine is $O(log\text{ }n)$, in a list of 8 items, we break it down 3 times to get a total of 8 list with only 1 item each.

The second routine is $O(n)$ because we need to iterate over the entire lists to merge them into an unique one.

**Combined both routines makes the time complexity of the whole algorithm to be $O(n\text{ }log\text{ }n)$.**

For sorting multiples type of data this is in fact very efficient. There are other sorting algorithms that only handles numbers and are more efficients but in terms of beign able to manage different types of data, this algorithm is the fastest.
