## Important concepts

The condition that makes my function to stop iterating it's called 'Base Case'.



If the condition case isn't reach yet, then the result of the operation it's called a 'Recursive Case'.




If the function isn't instructed to ask for a base case, it is resulted in a _theoretically_ infinite recurse case, and that is known as <b>stack overflow</b>.

Also, even if I have defined the base case, I need to be sure that the condition to reach the base case is feasible (e.g. if my condition is 1 > 2, it will never be reached thus I will have a stack overflow).

As well as be sure that the condition case is feasible, also I need to check that my function is getting out (return something) if my base case is reached, otherwise, I will still have a stack overflow because it will keep calling to itself.




## Call stack

A call stack is a data structure that keeps track of the execution of a program. It stores information about the active subroutines or functions (methods in the case of object-oriented programming) in a program, including the order in which they were called and their parameters and local variables. The call stack follows the Last In, First Out (LIFO) principle, meaning that the last function called is the first one to finish and be removed from the stack.



In the given code, I only call my method 'funcOne.' However, the first function on the call stack is 'funcThree.' This might seem counterintuitive, but it happens because the call stack prioritizes the last function called as the first item. In this case, 'funcOne' initially calls 'funcTwo,' and 'funcTwo' further calls 'funcThree.' Therefore, in terms of the call stack, 'funcOne' is initially at the top, then it shifts down when 'funcTwo' is called, and finally, all existing elements in the call stack shift down again when 'funcTwo' calls 'funcThree.'



## Understanding recursion with factorials.

Factorials provide a helpful context for grasping the concept of recursion. They are a mathematical technique enabling the calculation of extensive multiplication sequences using a single expression. For example, when calculating $$4!$$ , you break it down into $$ 4 * 3!$$, then $$ 4 * 3 * 2!$$, and ultimately $$ 4 * 3 * 2 * 1!$$ Here,  1! acts as the base case, and the recursion concludes once this base case is reached. This recursive breakdown illustrates how a complex computation, like factorials, can be systematically addressed through step-by-step simplifications.
